```python
import pandas as pd
import numpy as np
```

```python
#Salvando o conteudo do csv dentro de um objeto dataframe do panda
dataframe = pd.read_csv('train.csv')
dataframe.head()
```

```python
#listando as colunas do dataframe
dataframe.columns
```

```python
#Criando um objeto apenas com as colunas necessárias
df = dataframe[['NU_INSCRICAO','NU_NOTA_CN','NU_NOTA_CH', 'NU_NOTA_LC','NU_NOTA_MT', 'NU_NOTA_REDACAO']]
```

```python
df.columns
```

```python
df.shape
```

```python
#Apagando as linhas que contem qualquer valor nulo
dftratado = df.dropna(how='any')
```

```python
dftratado.describe()
```

```python
#criando objetos com peso das provas
pMatematica = 3.0
pCienciasNatureza = 2.0
pLiguagemCodigo = 1.5
pCienciasHumanas = 1.0
pRedacao = 3.0
```

```python
#percorrenco o dataframe e calculando a media ponderada
data = []
for indice, aluno in dftratado.iterrows():
    CH = float("{0:.2f}".format(aluno.NU_NOTA_CH))
    CN = float("{0:.2f}".format(aluno.NU_NOTA_CN))
    LC = float("{0:.2f}".format(aluno.NU_NOTA_LC))
    MT = float("{0:.2f}".format(aluno.NU_NOTA_MT))
    RD = float("{0:.2f}".format(aluno.NU_NOTA_REDACAO))
    
    #print( ((CH * pCienciasHumanas) + (CN * pCienciasNatureza) +  (LC * pLiguagemCodigo) + (MT * pMatematica) + (RD * pRedacao)) /  float(10.5))
    #print(type(aluno.NU_NOTA_REDACAO))
    
    
    mp =  ((CH * pCienciasHumanas) + (CN * pCienciasNatureza) +  (LC * pLiguagemCodigo) + (MT * pMatematica) + (RD * pRedacao)) / \
    (pMatematica + pCienciasNatureza + pLiguagemCodigo + pCienciasHumanas + pRedacao) 
    
    print(float("{0:.2f}".format(mp)))
    
    data.append([aluno.NU_INSCRICAO, float("{0:.2f}".format(mp))])  
```

```python
dfinal = pd.DataFrame(data, columns=['NU_INSCRICAO', 'NOTA_FINAL'])
```

```python
#ordenando pela nota final, decrescente para obter os primeiros colocados
dfinal = dfinal.sort_values(['NOTA_FINAL'], ascending=False)
```
